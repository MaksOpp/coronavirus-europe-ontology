```python
import pandas as pd
```

```python
covid_data = pd.read_csv('owid-covid-data.csv')
```

```python
covid_data.head(5)
```

```python
covid_data.columns
```

```python
covid_data = covid_data[covid_data['continent']=='Europe']
```

```python
covid_data = covid_data[['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'population', 'population_density', 'median_age', 'gdp_per_capita', 'cvd_death_rate']]
```

```python
dem_data = pd.read_csv('demographic-data.csv', sep=';')
```

```python
dem_data.head(5)
```

```python
data = pd.merge(covid_data, dem_data, left_on="location", right_on="Country")
```

```python
data.to_csv('covid_merged_data_clean.csv')
```
