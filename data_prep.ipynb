```python
import pandas as pd
```

```python
covid_data = pd.read_csv('csv/covid_data.csv')
```

```python
covid_data.head(5)
```

```python
covid_data.columns
```

```python
covid_data = covid_data[covid_data['continent']=='Europe']
```

```python
covid_data = covid_data[['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'population', 'population_density', 'median_age', 'gdp_per_capita', 'cvd_death_rate']]
```

```python
dem_data = pd.read_csv('csv/demographic_data.csv', sep=';')
```

```python
dem_data.head(5)
```

```python
data = pd.merge(covid_data, dem_data, left_on="location", right_on="Country")
data = data.dropna()
data.reset_index(drop=True, inplace=True)
```

```python
data = data.rename(columns={"Country": "country", "Overall Life Expectancy": "overall_life_expectancy", "Female Life Expectancy": "female_life_expectancy", "Male Life Expectancy": "male_life_expectancy", "Number of doctors per 1000 people": "number_of_doctors_per_1000_people"})
data = data.drop(columns=['country'])
data = data.rename(columns={"location": "country"})
```

```python
data['number_of_doctors_per_1000_people'] = data['number_of_doctors_per_1000_people'].apply(lambda x: float(x.replace(',','.')))
```
